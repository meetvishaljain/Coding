```python
class BankAccount():
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
        
    def deposit(self,amount):
        self.balance = self.balance+amount
        print("Rs.{} has been deposited to {} Account".format(amount,self.owner))
        
    def withdrawl(self,amount):
        if amount<=self.balance:
            print("Rs.{} has been withdrawn from {} Account".format(amount,self.owner))
            self.balance = self.balance-amount
        else:
            print("Insufficient Funds!!")
```

```python
myaccount = BankAccount('Vishal',500)
```

```python
myaccount.balance
```

```python
myaccount.deposit(400)
```

```python
myaccount.balance
```

```python
myaccount.withdrawl(200)
```
