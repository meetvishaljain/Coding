```python
class Line():
    def __init__(self,coord1,coord2):
        self.coord1=coord1
        self.coord2=coord2
        
    def distance(self):
        x1,y1=self.coord1[0],self.coord1[1]
        x2,y2=self.coord2[0],self.coord2[1]
        #print(x1,y1,x2,y2)
        return ((y2-y1)**2-(x2-x1)**2)**0.5
    
```

```python
c1=(5,10)
c2=(8,9)
L1 = Line(c1,c2)
```

```python
L1.distance()
```
